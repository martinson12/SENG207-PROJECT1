#### ALL NULL VALUES MUST BE REPLACED WITH THE MEAN OF EACH COLUMN, AND ALL DUPLICATES MUST BE REMOVED.

<p>By using the fillna() function, we can replace null values with the corresponding column's mean, median, or mode. Using the mean value and the drop_duplicate() method to remove all duplicates from the dataset is a common strategy that is helpful when there are only a few null values in the dataset, as is the case with this dataset.</p>

```python
dataset = dataset.fillna(meanDATA).drop_duplicates()
```

### ASSIGNING EACH COLUMN'S MEAN TO VARIABLES

```python
meanFA = dataset.fixed_acidity.mean()
meanVA = dataset.volatile_acidity.mean()
meanCA = dataset.citric_acid.mean()
meanRS = dataset.residual_sugar.mean()
meanC = dataset.chlorides.mean()
meanFSD = dataset.free_sulfur_dioxide.mean()
meanTSD = dataset.total_sulfur_dioxide.mean()
meanD = dataset.density.mean()
meanPH = dataset.pH.mean()
meanS = dataset.sulphates.mean()
meanA = dataset.alcohol.mean()
meanQ = dataset.quality.mean()
indexing = dataset.index
fa = dataset.fixed_acidity
```

### GRAPHING THE DATA


A few visual inferences that are concealed can be seen in the graphs. The dataset's distributions, outlines, and connections between variables are displayed in the dataset's bar graphs, pie charts, and histograms.



```python
# data for pie chart
# total columns excluding ID is 12
colors = ['green', 'orange', 'red', 'violet', 'blue', 'purple',
          'pink', 'brown', 'gray', 'black', 'white', 'beige']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
meanValues = [meanFA, meanVA, meanCA, meanRS, meanC,
              meanFSD, meanTSD, meanD, meanPH, meanS, meanA, meanQ]
meanTopics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
xtitle = 'Column names'
ytitle = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
                'volatile_acidity (VA)',
                'citric_acid (CA)',
                'residual_sugar (RS)',
                'chlorides (C)',
                'free_sulfur_dioxide (FSD)',
                'total_sulfur_dioxide (TSD)',
                'density',
                'pH',
                'sulphates',
                'alcohol',
                'quality']
```

```python
labels = meanTopics
sizes = meanValues
```

#### USING A PIE CHART TO COMPARE THE MEAN VALUES OF THE DATA


```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index',  loc='upper left',)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```

### Using a bar graph, compare the data's means.



```python
# plotting a bar graph

fig = plt.figure(figsize=(10, 5))

plt.bar(labels, sizes, width=0.5, color=colors)
plt.xlabel(xtitle)
plt.ylabel(ytitle)
plt.title('Comparing mean of columns in a bar chart')
plt.legend(loc='upper right', )
plt.show()
```
